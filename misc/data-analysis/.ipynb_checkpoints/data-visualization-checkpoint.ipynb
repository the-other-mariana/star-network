```python
import math
def bar_data(bars, nums, accumulate, x_vals, floats=True, fs=1):
    num_ticks = 5
    intervals = np.linspace(x_vals[0], x_vals[1], bars)
    intervals = list(intervals)
    intervals = [0.0] + intervals

    heights = [0 for i in intervals[1:]]
    xs = list(range(len(heights)))
    xlabels = []
    if floats:
        xlabels = ["[{:.2f}-{:.2f})".format(intervals[i-1], intervals[i]) for i in range(1, len(intervals))]
    else:
        for i in range(1, len(intervals)):
            if i % num_ticks == 0:
                xlabels.append(f"{(intervals[i-1]/1000.0):.1f}k")
            else:
                xlabels.append('')
    for i in range(len(intervals) - 1):
        minVal = intervals[i]
        maxVal = intervals[i + 1]
        for n in range(len(nums)):
            if nums[n] >= minVal and nums[n] < maxVal:
                heights[i] += 1
    # average
    for i in range(len(intervals) - 1):
        heights[i] = heights[i] / fs
    return xs, intervals, heights, xlabels
```

```python
ns = []
frames = list(TYPES.keys())
for f in range(len(frames)):
    n = list(db[db['FRAME_TYPE'] == frames[f]]['PCKT_NUM'].values)
    n = max(n)
    ns.append(n)
min(ns)
```

```python
import scipy
tests = list(db['TEST_TYPE'].unique())
frames = list(TYPES.keys())

figt = plt.figure( figsize=(10,18) )
rows = len(tests)
axes = figt.subplots(rows, 1)
bars = 30

for t in range(len(tests)):
    for f in range(len(frames)):
        df = db[(db['FRAME_TYPE'] == frames[f]) & (db['TEST_TYPE'] == tests[t])][['TIME(MS)', 'PCKT_NUM']]
        nums = list(df['PCKT_NUM'].values)
        times = list(df['TIME(MS)'].values)
        # from overall db
        x_vals = list(db['PCKT_NUM'].values)
        xs, intervals, heights, xlabels = bar_data(bars, nums, times, (0, min(ns)), False)
        axes[t].set_title(f'Number of Packets, Test={tests[t]}')
        if t == (len(tests)-1):
            axes[t].set_xticks(xs)
            axes[t].set_xticklabels(xlabels, rotation=0)
        else:
            axes[t].set_xticks([])
            axes[t].set_xticklabels([])
        lab = frames[f]
        if t > 0:
            lab = "_nolegend_"
        xnew = np.linspace(min(xs), max(xs), 5000) 
        spl = scipy.interpolate.make_interp_spline(xs, heights, k=3)  # type: BSpline
        power_smooth = spl(xnew)
        axes[t].plot(xnew, power_smooth, color=colors2[f], label=lab, lw=5)
        axes[t].yaxis.grid(True)
        #axes[t].plot(np.array(xs) + (offset[f]), heights, color=colors[f], linestyle='-')
        #axes[t].bar(np.array(xs) + (offset[f]), heights, width=step, color=colors[f], label=lab)
        
figt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)
figt.tight_layout()
plt.savefig(f"packets-per-test-ts-smooth.png", dpi=500)
plt.show()
```

```python
from scipy.interpolate import make_interp_spline, BSpline
from scipy.interpolate import interp1d
from scipy.stats import poisson
from scipy.special import factorial

tests = list(db['TEST_TYPE'].unique())
frames = list(TYPES.keys())
e = 2.71828

figt = plt.figure( figsize=(5,2) )
rows = len(tests)
axes = figt.subplots(1, 1)
bars = 30

periods = {'DATA': 150, 'CMD':75, 'ACK': 120, 'BCN':1.15}

for f in range(len(frames)):
    nums = []
    times = []
    for t in range(len(tests)):
        df = db[(db['FRAME_TYPE'] == frames[f]) & (db['TEST_TYPE'] == tests[t])][['TIME(MS)', 'PCKT_NUM']]
        nums += list(df['PCKT_NUM'].values)
        times += list(df['TIME(MS)'].values)
        # from overall db
        x_vals = list(db['PCKT_NUM'].values)
    #axes[0].set_xticks([])
    axes.set_xticks(xs)
    axes.set_xticklabels(xlabels, rotation=0)
    if t > 0:
        lab = "_nolegend_"
    xs, intervals, heights, xlabels = bar_data(bars, nums, times, (0, min(ns)), False, len(tests))
    #axes[0].set_title(f'Average Number of Packets Through Time Per FCF Type')
    axes.set_title(f'Probability Distribution Function Fitting')
    #axes[0].yaxis.grid(True)
    axes.xaxis.grid(False)
    axes.yaxis.grid(False)
    #axes.yaxis.grid(True)
    #axes[0].plot(np.array(xs), heights, color=colors[f], linestyle='-', label=frames[f])
    
    if frames[f] == 'DATA' or frames[f] == 'CMD' or frames[f] == 'ACK':
        xnew = np.linspace(min(xs), max(xs), 1000)
        x = np.linspace(0, min(ns), 1000)
        period = periods[frames[f]]
        l = 1.0 / period
        expo = sorted(heights, reverse=True)[1]*(1 - e**(-l*x))
        axes.plot(xnew, expo, color=colors2[f], lw=5)
        axes.plot(np.array(xs), heights, color=colors[f], linestyle='-', label=frames[f])
    if frames[f] == 'BCN':
        xnew = np.linspace(min(xs), max(xs), 1000)
        mu = periods[frames[f]]
        amp = sorted(heights, reverse=True)[0]
        p = amp*2*np.exp(-mu)*np.power(mu, xnew)/factorial(xnew)
        axes.plot(xnew, p, color=colors2[f], lw=5)
        axes.plot(np.array(xs), heights, color=colors[f], linestyle='-', label=frames[f])
    #axes.bar(np.array(xs) + (offset[f]), heights, width=step, color=colors[f], label=frames[f])
    
#box = axes[0].get_position()
#axes[0].set_position([box.x0, box.y0, box.width * 0.8, box.height])
box = axes.get_position()
axes.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
#axes[0].legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
axes.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
figt.tight_layout()
plt.savefig(f"packets-ts-avg-expo.png", dpi=500)
plt.show()
```

```python
test_types = ['Control', '802.15.4', 'RFs', 'Home Act.', 'Wind', 'Wifi']
df = db[['TIME(MS)','LENGTH', 'RSSI','CRC_OK','TEST_TYPE']]
mean_rssi = int(df['RSSI'].mean())
df['RSSI'] = df['RSSI'].fillna(mean_rssi)
cont = ['TIME(MS)','LENGTH', 'RSSI']
fig = plt.figure( figsize=(7, 2) )
axes = fig.subplots(1,3)

flierprops = dict(marker='D', markerfacecolor='mediumturquoise', markersize=6,  markeredgecolor='gray', alpha=.25)
sns.set_style("whitegrid")
sns.color_palette("flare", as_cmap=True)

for c in cont:
    x = df[c].values
    mu = x.mean()
    stdev = x.std()
    st = (x-mu)/stdev
    idx = cont.index(c)
    print(c, idx, mu, stdev)
    axes[idx] = sns.boxplot(x=st, y=df['TEST_TYPE'], ax=axes[idx], palette="Spectral", flierprops=flierprops)
    axes[idx].set_xticks([])
    axes[idx].set_xlabel(c, fontsize=10)
    axes[idx].set_ylabel('')
    axes[idx].set_yticklabels([])
    if idx == 0:
        axes[idx].set_ylabel('Interference Type',fontsize=12)
        axes[idx].set_yticklabels(test_types, fontsize=10)

fig.tight_layout()
plt.subplots_adjust(hspace = 0.0)
plt.savefig(f"box-plots.png", dpi=500)
plt.show()
```
